```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Position_Salaries.csv')
```

```python
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

#### Fitting decision Tree Regressor to the dataset

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(X,y)
```

#### Predicting the new result

```python
y_pred = dtr.predict([[6.5]])
y_pred
```

#### Visualising the Decision Tree Results

```python
plt.scatter(X,y,c='b')
plt.plot(X, dtr.predict(X), c='grey')
plt.title('Decision Tree Regression Tree results')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

But it is not proper because irrespective of the graph the Regressor is non continious one.

#### Visualising the Decision Tree Results in higher resolution

```python
X_grid = np.arange(min(X),max(X),0.01)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X,y,c='b')
plt.plot(X_grid, dtr.predict(X_grid), c='grey')
plt.title('Decision Tree Regression Tree results')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```
