```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style = 'white', palette = 'BrBG_r')
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,[3,4]].values
```

```python
x[:4]
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(16,8))
dendogram = sch.dendrogram(sch.linkage(x,method = 'ward'))
plt.title('Dendogram',  size=32)
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc = hc.fit_predict(x)
y_hc
```

```python
plt.figure(figsize=(14,8))
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],s=100,c='r',label='Carefull')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],s=100,c='b',label='Standard')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],s=100,c='g',label='Target')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],s=100,c='y',label='Careless')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],s=100,c='orange',label='Sensible')
plt.title('Clusters of clients')
plt.xlabel('Annual Income($)')
plt.ylabel('Spending Score(1-100)')
plt.legend(bbox_to_anchor=(1.12,1))
```
